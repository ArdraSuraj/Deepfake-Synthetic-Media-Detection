```python
import pandas as pd

dataset_path = "/kaggle/input/datasets/payaldhokane/deepfake-and-synthetic-media-detection-dataset"
file_path = dataset_path + "/deepfake_detection_metadata_dataset.csv"

df = pd.read_csv(file_path)

print("First 5 rows:")
print(df.head())
```

```python
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
df = df.dropna()   # safest option
print("After removing missing values:", df.shape)
```

```python
print("\nColumns available:", df.columns)
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Remove missing values
df = df.dropna()

print("Final shape after cleaning:", df.shape)
```

```python
print("Label counts:")
print(df['label'].value_counts())

print("\nLabel percentage:")
print(df['label'].value_counts(normalize=True) * 100)
```

```python
print("\nMedia Type Distribution:")
print(df['media_type'].value_counts())
```

```python
print("\nGeneration Method Distribution:")
print(df['generation_method'].value_counts())
```

```python
print("\nNumerical Features Summary:")
print(df.describe())
```

```python
print("\nAverage scores by label:")
print(df.groupby('label')[
    ['lip_sync_score', 'visual_artifacts_score', 
     'lighting_inconsistency_score']
].mean())
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
label_counts = df['label'].value_counts()
print("✅ Exact Label Counts:\n")
print(label_counts)
```

```python
label_percent = df['label'].value_counts(normalize=True) * 100

print("\n✅ Exact Label Percentage (%):\n")
print(label_percent.round(2))
```

```python
total = len(df)

print("\n✅ Total Clean Records:", total)
```

```python
feature_means = df.groupby('label')[[
    'lip_sync_score',
    'visual_artifacts_score',
    'lighting_inconsistency_score'
]].mean()

print("\n✅ Accurate Feature Means (by Label):\n")
print(feature_means.round(4))
```

```python
print("\n✅ FINAL SUMMARY")
print("="*40)

print("\nTotal Records:", len(df))

print("\nLabel Counts:\n", df['label'].value_counts())

print("\nLabel %:\n", (df['label'].value_counts(normalize=True)*100).round(2))

print("\nFeature Means:\n", df.groupby('label')[[
    'lip_sync_score',
    'visual_artifacts_score',
    'lighting_inconsistency_score'
]].mean().round(4))
```
